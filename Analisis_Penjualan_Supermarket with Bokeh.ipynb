Import Library

```python
import pandas as pd
import numpy as np

from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange, DatetimeTickFormatter
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.models import NumeralTickFormatter
from bokeh.palettes import Spectral3,Spectral4,Spectral5
from bokeh.transform import cumsum
from math import pi
```

Membaca data yang dimiliki

```python
df=pd.read_csv('https://raw.githubusercontent.com/irwanmaulana42/google-collab/master/supermarket_sales.csv')
df.head()
```

```python
df.info()
```

Menambahkan kolom 'Bulan' pada tabel

```python
df['Date'] = pd.to_datetime(df.Date)
df['Bulan'] = pd.to_datetime(df['Date']).dt.month
df['Tahun'] = pd.to_datetime(df['Date']).dt.year
df.head()
```

Isi

Pertanyaan 1: Pada bulan apa penjualan terbanyak?


```python
df['Bulan'].value_counts()
```

```python
hasil = df.groupby('Bulan').sum()
```

```python
output_notebook()

# Convert 'bulan' to string representation
bulan = [str(Bulan) for Bulan, df in df.groupby('Bulan')]

source = ColumnDataSource(data=dict(bulan=bulan, total=hasil['Total']))

p = figure(x_range=FactorRange(*bulan), height=350, title="Total Penjualan per Bulan",
           toolbar_location=None, tools="")


p.vbar(x='bulan', top='total', width=0.9, source=source, color="steelblue")


p.y_range.start = 0
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Bulan"
p.yaxis.axis_label = "Total dalam USD ($)"
p.xaxis.major_label_orientation = 1.2
p.yaxis.formatter = NumeralTickFormatter(format="0")

show(p)
```

```python
hasil['Total']
```

Berdasarkan hasil diatas dapat dilihat transaksi terbanyak ada pada bulan Januari dan terbanyak kedua ada pada bulan Maret. Pada bulan januari total pendapatannya mencapai 116291.868 USD

```python
total=df.groupby(['Payment']).sum().sort_values('Payment',ascending=True).head(10)
total=total[['Total','gross income']].round(2)
total
```

```python
output_notebook()

# Group by 'Product line' and 'Payment' to count occurrences
grouped = df.groupby(['Product line', 'Payment']).size().reset_index(name='counts')

product_lines = grouped['Product line'].unique()
payments = grouped['Payment'].unique()

# Creating a nested list of tuples for FactorRange
y = [(product_line, payment) for product_line in product_lines for payment in payments]
counts = [grouped.loc[(grouped['Product line'] == product_line) & (grouped['Payment'] == payment), 'counts'].values[0] for product_line in product_lines for payment in payments]

source = ColumnDataSource(data=dict(y=y, counts=counts))

p = figure(y_range=FactorRange(*y), height=800, title='Plot Hitung untuk menunjukan jenis pembayaran pada tiap jenis produk',
           toolbar_location=None, tools="")

p.hbar(y='y', right='counts', height=0.9, source=source, line_color="white",
       fill_color=factor_cmap('y', palette=Spectral4, factors=payments, start=1, end=2))

p.x_range.start = 0
p.ygrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

p.yaxis.axis_label = "Jenis Produk"
p.xaxis.axis_label = "Jumlah"

show(p)
```

```python
output_notebook()

# Group by 'Payment' and 'City' to count occurrences
grouped = df.groupby(['Payment', 'City']).size().reset_index(name='counts')

payments = grouped['Payment'].unique()
cities = grouped['City'].unique()

# Creating a nested list of tuples for FactorRange
x = [(payment, city) for payment in payments for city in cities]
counts = [grouped.loc[(grouped['Payment'] == payment) & (grouped['City'] == city), 'counts'].values[0] if (payment, city) in grouped[['Payment', 'City']].values else 0 for payment in payments for city in cities]

source = ColumnDataSource(data=dict(x=x, counts=counts))

p = figure(x_range=FactorRange(*x), height=350, title='Menunjukan metode pembayaran terbanyak diberbagai cabang',
           toolbar_location=None, tools="")

p.vbar(x='x', top='counts', width=0.9, source=source, line_color="white",
       fill_color=factor_cmap('x', palette=Spectral4, factors=cities, start=1, end=2))

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xaxis.group_text_font_size = '10pt'
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

p.xaxis.axis_label = "Metode Pembayaran"
p.yaxis.axis_label = "Jumlah"

show(p)
```

Pertanyaan 4: Apakah Member melakukan pembelian lebih banyak daripada Non Member?

```python
output_notebook()

# Calculate value counts
customer_counts = df['Customer type'].value_counts()
customer_data = pd.DataFrame({'Customer type': customer_counts.index, 'value': customer_counts.values})

# Calculate the cumulative sum for the angle data
customer_data['angle'] = customer_data['value']/customer_data['value'].sum() * 2 * pi

# Assign colors
customer_data['color'] = Spectral4[:len(customer_counts)]

p = figure(height=350, title="Pie Chart for Customer Type", toolbar_location=None,
           tools="hover", tooltips="@{Customer type}: @{value}", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='Customer type', source=customer_data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

```python
df['Customer type'].value_counts()
```

```python
output_notebook()

# Group the data by 'City' and 'Customer type'
grouped = df.groupby(['City', 'Customer type']).size().reset_index(name='counts')

cities = grouped['City'].unique()
customer_types = grouped['Customer type'].unique()

# Prepare the data for the grouped bars
x = [(city, ctype) for city in cities for ctype in customer_types]
counts = [grouped.loc[(grouped['City'] == city) & (grouped['Customer type'] == ctype), 'counts'].values[0] for city in cities for ctype in customer_types]

source = ColumnDataSource(data=dict(x=x, counts=counts))

p = figure(x_range=FactorRange(*x), height=350, title='Menunjukan tipe pembeli di berbagai cabang',
           toolbar_location=None, tools="")

p.vbar(x='x', top='counts', width=0.9, source=source, line_color="white",
       fill_color=factor_cmap('x', palette=Spectral4, factors=customer_types, start=1, end=2))

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1.0
p.xaxis.group_text_font_size = '10pt'
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

p.xaxis.axis_label = "City"
p.yaxis.axis_label = "Count"

show(p)
```

Pertanyaan 5: Apa Produk yang paling banyak dibeli?

```python
from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool, DatetimeTickFormatter
from bokeh.plotting import figure

output_notebook()

# Aggregate the quantity for each date
agg_df = df.groupby('Date').agg({'Quantity': 'sum'}).reset_index()

source = ColumnDataSource(data=dict(date=agg_df.Date, quantity=agg_df['Quantity']))

p = figure(width=1200, height=350, x_axis_type="datetime", title='Diagram Garis untuk menunjukan Penjualan perhari di Supermarket')

# Main line
p.line(x='date', y='quantity', source=source, line_width=3, color="steelblue", alpha=0.8, legend_label="Quantity")



# Add hover tool
hover = HoverTool()
hover.tooltips = [("Date", "@date{%F}"), ("Quantity", "@quantity")]
hover.formatters = {'@date': 'datetime'}
p.add_tools(hover)

# Formatting datetime axis for clarity
p.xaxis.formatter = DatetimeTickFormatter(days=["%b %d"], months=["%b %Y"], years=["%Y"])

p.y_range.start = 0
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Quantity'
p.title.text_font_size = '20pt'
p.legend.location = "top_left"

show(p)
```

5. Di kota Naypitaw produk yang paling banyak dibeli adalah Food and Beverages dan Fashion Accessories, di kota Yangon yang terbanyak dibeli adalah Home and Lifestyle dan Electronic Accessories, sedangkan di kota Mandalay yang terbanyak adalah Fashion Accessories dan Sports and Travel
